```python
# recursive call을 활용한 binary search
def binary_search_recursive(A, key, low, high):  # 리스트 A에서 key를 low부터 high 범위까지 재귀로 찾는 함수
    if low > high:  # 탐색 범위가 없어지면
        return -1  # 찾는 값이 없다는 뜻으로 -1 반환

    mid = (low + high) // 2  # 가운데 인덱스를 구함

    if A[mid] == key:  # 가운데 값이 찾는 값과 같으면
        return mid  # 해당 인덱스를 반환
    elif A[mid] > key:  # 가운데 값이 찾는 값보다 크면
        return binary_search_recursive(A, key, low, mid - 1)  # 왼쪽 범위에서 다시 탐색
    else:  # 가운데 값이 찾는 값보다 작으면
        return binary_search_recursive(A, key, mid + 1, high)  # 오른쪽 범위에서 다시 탐색
```
