```python
def is_prime_number(N):
    if N <= 1:
        return False
    is_prime = True
    for i in range(2, N):
        if N % i == 0:
            is_prime = False
            break
    return is_prime
```

```python
for i in range(1, 5):
    print(is_prime_number(i))
```
